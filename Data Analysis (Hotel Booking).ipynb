### **Data Analysis and Visualizations**

**Calculating the percentage of canceled and non-canceled bookings**

```python
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),color="teal",edgecolor="k",width=0.6)
plt.show()
```

**Visualize the Reservation status across different hotels**

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df,palette="bright")
legend_labels, _ =ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('no of reservations')
plt.legend(['Not canceled','Canceled'])
plt.show()
```

**Calculates the percentage of canceled and non-canceled bookings for that Resort Hotel.**

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

In **resort hotels**, most bookings, about **72%**, aren't canceled. However, around **27.9%** of reservations are getting canceled, which is still quite a sizable portion.

**Calculates the percentage of canceled and non-canceled bookings for that City Hotel.**

```python
City_hotel=df[df['hotel']=='City Hotel']
City_hotel['is_canceled'].value_counts(normalize=True)
```

**City hotels** are facing a high cancellation rate, with **41.7%** of reservations getting canceled, while only **58.2%** are staying confirmed.

**Calculating the average daily rate for Resort Hotel and City Hotel**

```python
resort_hotel= resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel= City_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

The observation of the orange line (representing city hotel Average Daily Rate) intersecting with the blue line (representing resort hotel Average Daily Rate) at various points along the graph indicates fluctuating price comparisons between the two types of hotels over time. This supports our hypothesis that city hotel prices generally remain lower compared to resort hotels, except during busy times like holidays and weekends when more people are booking rooms, which could cause prices to rise.

### Determining the Reservation status per month

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation status per month',size=20)
plt.xlabel('Month')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

- January experienced the peak cancellation rate, while August had the lowest cancellation rate.
- August recorded the highest number of reservations, while December and January had the lowest.

#### Calculated the average daily rate per month to assess price fluctuations in relation to cancellations

```python
plt.figure(figsize=(15,8))
plt.title("Average daily rate per month", fontsize=30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index(),palette='husl')
plt.show()
```

Investigating the data, we found that January showcased the highest cancellation rate and high price, while August showed the lowest cancellation rate and lower prices. Also, August was the busiest month for bookings, while December and January were slower. These findings suggest a clear relationship between cancellation rates and pricing dynamics  higher prices tend to coincide with increased cancellations, whereas lower prices correspond to decreased cancellation rates. This underscores the importance of pricing strategies in managing reservation cancellations effectively.

#### Determining the cancellation percentage across various countries.

```python
cancelled_data=df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 country with reservation canceled')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.show()
```

"Portugal (PRT) has the highest cancellation rate at 70.07%, significantly higher than the second highest rate of 6.25% in the United Kingdom (GRD). This stark difference suggests that Portugal should consider enhancing its facilities, adjusting its pricing strategies, offering more discounts, and running promotional campaigns to reduce the high cancellation rate.

Improving the customer experience through upgraded amenities and better services can make stays more appealing, thereby reducing cancellations. Additionally, revisiting pricing models to make them more competitive and attractive, coupled with targeted discounts and special offers, can incentivize more bookings and minimize the likelihood of cancellations. Effective marketing campaigns can further bolster these efforts by attracting a more committed customer base.

By addressing these areas, Portugal can work towards aligning its cancellation rates more closely with other countries, thereby stabilizing and potentially increasing its reservation rates."

### Analyzing Cancellation Rates by Market Segment

```python
cancelled_data['market_segment'].value_counts(normalize=True)
```

- Online TA constitutes 46.9% of the cancellations, indicating that a majority of cancellations come from online travel agencies.
  
**Recommendations**
- Address Online TA Cancellations: Since Online TA has the highest proportion of cancellations, consider reviewing and potentially tightening the cancellation policies associated with these bookings. Enhance customer engagement and support for online bookings to reduce cancellations.

```python

# Calculate the average daily rate for canceled bookings
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Calculate the average daily rate for non-canceled bookings
not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plot the average daily rates
plt.figure(figsize=(15,6))
plt.title('Average Daily Rate', fontsize=30)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')
plt.xlabel('Reservation Status Date')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend()
plt.show()
```

The chart indicates that cancellations occur more frequently when the average daily rate is higher compared to when it is lower. This clearly supports the previous analysis, demonstrating that increased prices lead to a higher rate of cancellations.
